```python
def F(x):
  return 4*x**2+3*x-6+300*np.sin(x)
x=np.linspace(-10,10,1000)
plt.figure(figsize=(6,3)) #設定繪圖區大小
#繪圖區的標題，設定用中文字體twfont1，字體大小20
plt.title("標題說明",fontproperties=twfont1,fontsize=20)
#設定橫軸和縱軸的標題
plt.xlabel("x軸標題",fontproperties=twfont1,fontsize=20)
plt.ylabel("y軸標題",fontproperties=twfont1,fontsize=20)
plt.plot(x,F(x),"b:",label="F(x)")
```

```python
x=np.linspace(200,450,200)
y=np.linspace(200,450,200)
x,y=np.meshgrid(x,y)
#設定繪圖區大小
plt.figure(figsize=(10,10))                
#繪圖區的標題，設定用中文字體twfont1，字體大小20
plt.title("等高線圖",fontproperties=twfont1,fontsize=20)
#設定橫軸和縱軸的標題
plt.xlabel("x",fontproperties=twfont1,fontsize=15)
plt.ylabel("y",fontproperties=twfont1,fontsize=15)
CS=plt.contour(x,y,(x-200)**2+(y-200)**2,cmap="cool")
plt.clabel(CS,inline=1,fontsize=10)
```
