## 載入pytorch相關套件

```python
import torch
#建立神經網路用
import torch.nn as nn
import torch.nn.functional as F
#載入優化器
from torch import optim
#預處理資料用
from torch.utils.data import Dataset,DataLoader
```

## 建立DataSet類別

```python
class MyDataset(Dataset): 
  def __init__(self,Train=True):
    data=np.loadtxt(folder+"資料集/BostonHouse.csv",delimiter=",")
    data_x=data[:,:13]
    data_x=(data_x-data_x.mean(axis=0))/data_x.std(axis=0)
    data_y=data[:,13].reshape((-1,1))
    if Train==True:
      data_x=data_x[:400]
      data_y=data_y[:400]
    else:
      data_x=data_x[400:]
      data_y=data_y[400:]
    self.data = torch.tensor(data_x)
    self.label = torch.tensor(data_y)
  def __getitem__(self, index): 
    return self.data[index], self.label[index]
  def __len__(self): 
    return len(self.data)
```

## 實體化訓練和測試資料集的DataSet、DataLoader

```python
TrainDS=MyDataset(Train=True)
TestDS=MyDataset(Train=False)
TrainDL=DataLoader(dataset=TrainDS,batch_size=100,shuffle=True)
TestDL=DataLoader(dataset=TestDS,batch_size=len(TestDS),shuffle=False)
```

```python
net=Net() 
optimizer = optim.SGD(net.parameters(), lr=0.008)
LossFun=
Step_L=[]
```

## 開始訓練神經網路

```python
net.train()
for i in range(40):
  for data,label in TrainDL:
    optimizer.zero_grad()
    Yh=net(data.float())
    loss=LossFun(Yh,label.float())
    Step_L.append(float(loss))
    loss.backward()
    optimizer.step()
  print("訓練回合：",i+1,"損失值：",loss)
  display.clear_output(wait=True)
```

## 觀察訓練過程中的損失函數Loss變化
